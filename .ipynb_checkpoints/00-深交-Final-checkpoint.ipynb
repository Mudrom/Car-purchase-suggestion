```python
# -*- coding: utf-8 -*-
import requests
import time
import random
import os
import json
import csv
import re
h1={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.71 Safari/537.36'}
#下载pdf公告的函数
def downloadpdf(url,h,file,t):
    resp = requests.get(url,headers=h,stream=t)
    f = open(file, 'wb')
    f.write(resp.content)
    f.close()

# 定义爬取函数
def get_pdf_address(pageNum,start_date,end_date):
    url='http://www.szse.cn/api/disc/announcement/annList?random=%s'% random.random()
    headers={'Accept': 'application/json, text/javascript, */*; q=0.01'
    ,'Accept-Encoding': 'gzip, deflate'
    ,'Accept-Language': 'zh-CN,zh;q=0.9'
    ,'Connection': 'keep-alive'
    ,'Content-Length': '127'#
    ,'Content-Type': 'application/json'
    ,'Host': 'www.szse.cn'
    ,'Origin': 'http://www.szse.cn'
    ,'Referer': 'http://www.szse.cn/disclosure/listed/notice/index.html'
    ,'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.71 Safari/537.36'
    ,'X-Request-Type': 'ajax'
    ,'X-Requested-With': 'XMLHttpRequest'}
    
    pagenum = int(pageNum)
    payload={'seDate': ["{}-10-26".format(start_date), "{}-11-09".format(end_date)] 
        ,'bigCategoryId': ["010301"]#
        ,'channelCode': ["listedNotice_disc"]
        ,'pageNum':pagenum
        ,'pageSize': '30'
        }
    r = requests.post(url,headers = headers,data = json.dumps(payload))
    result = r.json()
    return result

#新建csv文件，存储公告详细信息
csvf = open('D:\\0final\\深圳证券市场公告.csv', 'w', encoding='gbk', newline='')
writer = csv.writer(csvf)
writer.writerow(('公司名', '股票代码', '发布时间'))
count = 0
url_head = 'http://disc.static.szse.cn/download/'
end = 2021
while end >2020:
    start = end 
    result = get_pdf_address(1,start,end)
    print("开始爬取{}-10-26至{}-11-09的年报".format(start,end))
    pages = int(result['announceCount']/30 + 1)+1
    print("共%s页" % pages)
    for i in range(1,pages):
        print("爬取深交所年报下载地址第{}页".format(i))
        result = get_pdf_address(i,start,end)
        num = len(result['data'])
        for each in range(num):
            code = result['data'][each]['secCode'][0]
            name = result['data'][each]['secName'][0]
            title=result['data'][each]['title']
            p_time = result['data'][each]['publishTime'].split(' ')[0]
            url_d=result['data'][each]['attachPath']
            if re.search('年度报告',title,re.S):
                if re.search('摘要',title,re.S):###避免下载一些年报摘要等不需要的文件###
                    pass
                elif re.search('半',title,re.S):###避免下载一些年报摘要等不需要的文件###:
                    pass
                else:
                    adjunctUrl ='http://disc.static.szse.cn/download/'+url_d
                    pdffile ='D:\\0final\\pdf\\'+code+'-'+p_time+'-'+ title+'.pdf'
                    if re.search('ST',title,re.S):###下载前要将文件名中带*号的去掉，因为文件命名规则不能带*号，否则程序会中断###
                            pdffile='D:\\0final\\pdf\\'+code+'-'+p_time+'-'+'ST' +'.pdf'
                    downloadpdf(url=adjunctUrl,h=h1, file=pdffile,t=True)
                    writer.writerow(( code, name,p_time))
                    
                   # if re.search('ST',title,re.S):###下载前要将文件名中带*号的去掉，因为文件命名规则不能带*号，否则程序会中断###
                    #    pdffile='D:\\0F\\PDF\\'+code+'-'+'ST'+'-'+ p_time+'.pdf'
                    
                    print('获取完成')
                    print("========================================")
                    
                    time.sleep(random.random())
    end=end-1
    
```
